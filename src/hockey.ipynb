CF stands for Corsi For, which measures the total number of shot attempts a team makes while a specific player is on the ice. It is a foundational advanced statistic used as a proxy for puck possession. 

Definition: 
CF = Goals + Saved Shots + Missed Shots + Blocked Shots
CA = Corsi Against.  Same metric...Opponents CF
CF % = CF / CF + CA
Above 50% is good, Above 55% is elite

What is Included in CF?
CF counts every shot attempt directed at the opponent's net, regardless of whether it reaches the goalie: 

Shots on Goal: Attempts that hit the net and are either a goal or a save.
Missed Shots: Attempts that go wide or over the net.
Blocked Shots: Attempts that are stopped by an opposing skater before reaching the net. 

Key Related Metrics
Analysts often use variations of CF to get a clearer picture of a player's impact: 
Metric 	Full Name	Definition
CF	Corsi For	Raw count of shot attempts for a team/player while on ice.
CA	Corsi Against	Raw count of shot attempts by the opponent while on ice.
CF%	Corsi For Percentage	The percentage of total shot attempts that were for the player's team: 

CF% Rel	Relative Corsi For %	The difference in a team's CF% when a player is on the ice vs. when they are off the ice.

Why It Matters
Possession Proxy: Since most shots happen in the offensive zone, a high CF suggests your team has the puck more often than the opponent.

Predictive Power: CF% is generally considered a better predictor of future success than goal differential because shot attempts occur more frequently, providing a larger sample size.

Performance Benchmarking: A CF% above 50% means the team is out-shooting the opponent with that player on the ice; players above 55% are typically considered "elite". 

The NHL officially refers to Corsi as SAT (Shot Attempts)

FF stands for Fenwick For, an advanced hockey statistic that counts all unblocked shot attempts (goals, shots on net, and missed shots) a team or player generates while on the ice. It excludes blocked shots, acting as a measure of offensive pressure and puck possession that is often considered more indicative of quality scoring opportunities than total shot attempts (Corsi). 

Key Details About FF:
Similar to CF, but excludes Blocked Shots

Definition: 
FF = Goals + Saved Shots + Missed Shot
FA = Fenwick Against.  Same metric...Opponents FF
FF % = FF / FF + CA


Purpose: It evaluates which team is controlling the puck in the offensive zone, excluding attempts that were stopped by defenders.

Difference from Corsi: Corsi includes blocked shots, while Fenwick removes them, as they are often considered more a function of defensive skill rather than offensive creation.

FF% (Fenwick For Percentage): This is the percentage of total unblocked shots that a team/player takes, calculated as: 
 (Fenwick For / (Fenwick For + Fenwick Against)).

iFF (Individual Fenwick For): Specifically refers to the unblocked shot attempts taken by a single player. 

It is commonly used alongside FA (Fenwick Against) to analyze game-level performance.

xGF stands for Expected Goals For, an advanced hockey statistic measuring the total quality of scoring chances a team or player generates while on the ice. It calculates the probability of unblocked shots becoming goals based on factors like distance, angle, and shot type. Higher xGF indicates better offensive pressure. 

Key details about xGF:

Definition: xGF represents the total expected goal value of all unblocked shot attempts (Fenwick) a team or player generates.

xGF % = xGF / (xGF + xGA)


What it Measures: Instead of just counting shots (like Corsi or Fenwick), xGF weighs them by likelihood of scoring (e.g., a rebound from the slot has a higher xG value than a wrist shot from the blue line).

Purpose: It is a strong indicator of future performance, often better than actual goals scored at predicting future success because it highlights high-quality play-driving.

How it is Calculated: Analysts use statistical models (logistic regression or machine learning) on historical play-by-play data to assign a probability value between 0 and 1 to every shot. For example, a shot from a high-danger area might have an xG of .3 (30% chance of scoring), while a long shot might be .02 (2% chance of scoring).
 

Related Stats:
xGA (Expected Goals Against): The quality of chances given up.
xGF% (Expected Goals Percentage): The share of total expected goals (xGF / (xGF + xGA)) a team produces. A percentage above 50% generally indicates a positive, high-quality offensive performance.


PDO is a metric used to gauge a team's or player's "puck luck" by combining their shooting and saving efficiency. It is also known by the NHL as SPSV% (Shooting Plus Save Percentage). 

Calculation
PDO is calculated by adding a team's Shooting Percentage (S%) and its Save Percentage (SV%), typically during even-strength (5-on-5) play. 

Shooting Percentage (S%): The percentage of a team's shots on goal that result in goals.
Save Percentage (SV%): The percentage of shots on goal against a team that its goaltenders save.
Example: If a team has a 10% shooting percentage and a 91% save percentage, their PDO is 101. 

Understanding the Results
The fundamental principle of PDO is regression to the mean. Because every shot on goal in a league either results in a goal or a save, the league-wide average PDO is always exactly 100 (or 1.000). 

Above 100: Indicates a team may be getting "lucky" or performing above sustainable levels. According to some analysts, a PDO over 102 suggests a team is "not as good as they seem" and likely to regress.

Below 100: Indicates a team may be facing "bad luck" or underperforming. A PDO below 98 often suggests a team is actually better than their current results show.
Sustainability: Over a full season, most teams eventually converge toward 100 as extreme variations in shooting or goaltending normalize. 

Why is it called "PDO"?
Contrary to popular belief, PDO is not an acronym. It was named after the online handle of Brian King, a hockey enthusiast who popularized the metric on early internet forums. 


SRS stands for Simple Rating System, a team-level metric that rates teams based on their goal differential and the strength of their schedule. It is denominated in goals above or below the league average, where 0.0 is perfectly average. 

How SRS is Calculated
The SRS is the sum of two main components:
Average Goal Differential (AGD): The team's total goal differential (goals for minus goals against) divided by the number of games played.
Strength of Schedule (SOS): The average SRS rating of all the opponents a team has played. 

The basic formula is:

SRS = Avg Goal Differential + Strength of Schedule

Key Characteristics
Iterative Process: Because a team's SRS depends on their opponents' SRS (and vice versa), the values must be calculated through an iterative process until the numbers converge and stabilize.

Performance Comparison: Beating a high-rated team by a wide margin significantly boosts a team's SRS, while losing a close game to a top-tier opponent may hurt a team's rating less than losing to a poor team.

Predictive Use: Analysts often use SRS to predict the outcome of future matchups. For example, if Team A has an SRS of 1.5 and Team B has an SRS of 0.5, Team A would be expected to win by roughly 1 goal on a neutral site.

Normalization:
The rating is adjusted so that the league average always remains at zero.
